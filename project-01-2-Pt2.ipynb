# **1. Perceptron model**

### **Define what a single-neuron network is, explaining its components: inputs, weights, bias, activation function, and output.**


Una red neuronal de solo una neurona es conocida como perceptrón. El perceptrón se divide en dos.

- Una sola capa: "El perceptrón de una sola capa solo puede aprender patrones linealmente separables."(Banoula, 2024)
- Multicapa:"Los perceptrones multicapa pueden aprender sobre dos o más capas que tengan un mayor poder de procesamiento."(Banoula, 2024)

El perceptrón, siendo la red neuronal en su forma más simple, se compone de solo una entrada con la capacidad de procesar y producir una salida. El perceptrón se compone de:

- **Entradas**: Datos que se proporcionan al perceptrón
- **Pesos**: Para cada entrada, hay un peso. Los pesos ayudan a determinar la importancia de cada entrada en la predicción final.
- **Sesgos**: Margen que permite ajustar la salida del perceptrón.
- **Función de activación**: Función que toma como referencia las entradas y los sesgos para producir la salida del perceptrón
-**salida**:Resultado/predicción final del perceptrón después de la función de activación.




### **If the perceptron model will be used for regression, explain how it can predict a continuous output value based on input features.**

- Por medio de una combinación de los pesos de cada entrada. Dicha combinación se utiliza para la función de activación, donde, el resultado será un valor numérico donde es ideal para predecir valores continuos como lo son temperaturas y costos. Esto porque se toma en cuenta la combinación lineal de los pesos de cada entrada y no se es requerida una función de activación no lineal, si no, una activación lineal, esto porque se busca obtener un valor de manera continua.

### **If the perceptron model will be used for classification, explain how it uses a decision boundary to classify inputs into two categories (binary classification).**

- Al igual que la regresión se hace una combinación de los pesos con cada una de sus entradas, con la diferencia que la función de activación que se requiere es una no lineal, esto genera que la salida sea no continua, sino, discreta. Por medio de la función de activación seleccionada, se compara la salida con un valor de margen, donde si el valor de salida es mayor que dicho margen, entonces se clasifica en una categoría, y si no, entonces se clasifica en la otra categoría, donde dichas categorías son separadas en un plano.


# **2. Activation functions**

Depending on the use case application where the single-neuron network will be applied,explore different activation functions used by the perceptron model:

### **Regression: Linear activation function. Keep in mind that a perceptron model can only capture linear relationships between inputs and outputs.**

La regresión es una técnica de aprendizaje supervisado que tiene como objetivo predecir valores continuos, como el valor futuro de una casa, por ejemplo. Para modelos de regresión, las funciones lineales son más comunes, aunque también se pueden usar funciones no lineales dependiendo los datos. Si usamos una función de activación lineal en un perceptrón, este sólo podrá modelar relaciones lineales entre entradas y salidas, lo cual suena bien si eso es lo que queremos lograr. La función de activación “linear” es la más básica para nuestros códigos en Python. (Rasamoelina et al., 2020)


### **Classification: Sigmoid or Tanh.**

Por otro lado, las funciones no lineales como la sigmoide, tangente hiperbólica o ReLU (Rectified Linear Unit) son utilizadas en modelos de clasificación. La función sigmoide convierte cualquier valor de entrada en un número entre 0 y 1, lo cual es útil para representar probabilidades, pero tiene el problema de que los gradientes pueden desaparecer en redes profundas, dificultando el entrenamiento. La función tangente hiperbólica (tanh), similar a la sigmoide, transforma la entrada en un valor entre -1 y 1, lo que le permite representar mejor los datos al incluir valores negativos. Aunque tiene una pendiente más pronunciada que la sigmoide, y sufre el mismo problema de desvanecimiento de gradientes en redes profundas. Ambas funciones son ideales para tareas de clasificación o cuando se desea limitar la salida a un rango específico. (Rasamoelina et al., 2020)

## **Preparar los datos**

```python
x = df[['last_purchase_amount','age']] # Se define 'x' como las características (features) que serán utilizadas: 'last_purchase_amount' y 'age'.

y = df['Cluster_Id']# Se define 'y' como la variable objetivo (target), en este caso el 'Cluster_Id'.

print(x.shape)# Se imprime la forma (shape) de los datos de entrada.

y.shape# Se imprime la forma de la variable objetivo 'y'.
```

Se calcula el nuevo tamaño de X y Y.

```python
print(x_train.shape) # Se imprime la forma del conjunto de entrenamiento para verificar que los datos se separaron correctamente.
```

Imprime el tamaño de x_train

## **Separar los datos en train y test**

```python
#Separar la data en train y test
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=42, shuffle=True)
# El 80% de los datos se usa para entrenamiento (train) y el 20% para prueba (test).
```

No hay salida para esta celda, pero realiza el train_test_split

## **Definir la arquitectura de la red neuronal**

```python
# Definimos la arquitectura de la red
def get_model():
    # Creamos el modelo con una capa de entrada, una capa oculta con 'tanh', y una capa de salida con 'softmax'
    model = tf.keras.Sequential([
        tf.keras.layers.InputLayer(shape=(x_train.shape[1],)),  # Ajustar el tamaño de la entrada
        tf.keras.layers.Dense(10, activation='tanh'),  # Primera capa oculta con tanh
        tf.keras.layers.Dense(len(np.unique(y_train)), activation='softmax')  # Capa de salida multiclase
    ])
    return model  # Retornar el modelo

# Crear el modelo
model = get_model()
```

```python
#Imprimimos el resumen del modelo
model.summary()
```

Imprime el resumen del modelo

```python
# Compilar el modelo con optimizador adam y pérdida 'sparse_categorical_crossentropy' para clasificación multiclase
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
# Entrenar el modelo
history = model.fit(x_train, y_train, epochs=200, batch_size=20, validation_split=0.20)
```

Imprime la información de cada una de las iteraciones que realiza el modelo.

## **Evaluar el modelo en el conjunto de prueba**

```python
# Evaluamos el modelo con los datos de prueba, y nos entrega la perdida y el accuracy
loss, accuracy = model.evaluate(x_test, y_test)
#Imprimimos el accuracy con 4 valores decimales.
print(f"Test Accuracy: {accuracy:.4f}")
```

Imprime el accuracy de los datos de prueba.

## **Graficar**

```python
# Graficamos el comportamiento de los datos de train y test

plt.figure(figsize=(12,6))
plt.plot(history.history['accuracy']) # Gráfica de la exactitud del conjunto de entrenamiento.
plt.plot(history.history['val_accuracy']) # Gráfica de la exactitud del conjunto de validación.
plt.xlabel('epoch') # Etiqueta del eje x.
plt.ylabel('accuracy') # Etiqueta del eje y.
plt.legend(['train', 'val']) # Leyenda para distinguir entrenamiento y validación.
plt.grid(); # Agregar una cuadrícula para mayor claridad en la gráfica.
```

Sin embargo, también se aprecia una especia de oscilación en la línea de validación, lo que podría deberse a ligeras variaciones durante el proceso de validación en cada época.

Este gráfico muestra la pérdida (loss) en función de las épocas, se puede ver una disminución significativa en las primeras épocas, lo que indica que el modelo está ajustándose adecuadamente. La pérdida de ambos conjuntos, entrenamiento y validación, tiende a converger a valores bajos después de muchas épocas, lo cual es bueno e indica un buen ajuste. Sin embargo, las pequeñas fluctuaciones u oscilaciones en la línea de validación podrían indicar alguna variabilidad en la predicción del modelo, aunque en general el comportamiento es estable.

## **Matriz de confusión**

```python
# Importar las funciones para calcular y mostrar la matriz de confusión.
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

# Usar el modelo para predecir los clusters (clases) en el conjunto de prueba.
y_pred_nn = model.predict(x_test)

# Convertir las probabilidades predichas a clases (el índice con el valor más alto es la clase predicha).
y_pred_nn_classes = np.argmax(y_pred_nn, axis=1)

# Crear la matriz de confusión
c_matrix = confusion_matrix(y_test, y_pred_nn_classes)

# Mostrar la matriz de confusión
disp = ConfusionMatrixDisplay(confusion_matrix=c_matrix, display_labels=np.unique(y_test)) # Utilizar el color azul en el gráfico.
disp.plot(cmap=plt.cm.Blues) # Utilizar el color azul en el gráfico.
plt.title('Matriz de Confusión - Red Neuronal')# Título del gráfico.
plt.show() # Mostrar el gráfico.
```

La matriz de confusión nos permite ver la cantidad de aciertos y errores que tiene el modelo al momento de categorizar. El gráfico se lee de la siguiente manera: en el eje x tenemos la categoría a la cual el modelo cree que pertenece el dato, y en el eje y tenemos la categoría a la que realmente pertenece el dato. Así, la suma de los valores que encontramos en la diagonal principal representa el número de aciertos, y todo lo que se encuentra fuera de esta son errores que comete el modelo. En este caso, el error más grande que comete el modelo es etiquetar un dato que pertenece a la clase 0 como si perteneciera a la clase 3, lo cual ocurre 8 veces.

De los 200 valores de prueba que se utilizan, la suma total de errores es de 21, mientras que el modelo acierta en 179 ocasiones. Estos datos nos permitirán conocer las métricas de desempeño en el siguiente apartado.

## **Cálculo de métricas de desempeño**

```python
def get_performance_metrics(c_matrix): #Permite el acceso a los datos de la matriz de confusión para obtener las métricas de desempeño.
    tn, fp, fn, tp = c_matrix[0][0], c_matrix[0][1], c_matrix[1][0], c_matrix[1][1] #Extracción de los valores de la matriz de confusión.

    accuracy = (tp + tn) / (tp + tn + fp + fn) #Cálculo de exactitud.
    precision = tp / (tp + fp) #Cálculo de presición.
    recall = tp / (tp + fn) #Cálculo de sensibilidad.
    specificity = tn / (tn + fp) #Especificidad.
    f1_score = 2 * (precision * recall)/(precision + recall) #Cálculo de F1-score.
    return accuracy, precision, recall, specificity, f1_score #Regresa los valores previamente calculados.

accuracy, precision, recall, specificity, f1_score = get_performance_metrics(c_matrix) #Obtener cálculo de las métricas.

print(f"Accuracy: {accuracy:0.6f}") #Desplegar Accuracy.(Exactitud)
print(f"Precision: {precision:0.6f}") #Desplegar Presicion.
print(f"Recall: {recall:0.6f}") #Desplegar Recall. (Sensibilidad)
print(f"Specificity: {specificity:0.6f}")  #Desplegar Specificity. (Especificidad)
print(f"F1-score: {f1_score:0.6f}") #Desplegar F1-Score. ()
```

Las métricas de desempeño obtenidas muestran que nuestro modelo es bastante preciso. Con una accuracy del 95.35%, el modelo clasifica correctamente la mayoría de los ejemplos. La precisión del 100% indica que todos los resultados positivos predichos por el modelo fueron correctos. El recall (sensibilidad) del 90.48% muestra que el modelo identifica bien los casos positivos, aunque sí podría mejorar un poco. La especificidad del 100% refleja que el modelo clasifica correctamente todos los ejemplos negativos. Finalmente, el F1-score de 0.95 combina precisión y recall, indicando que el modelo tiene un buen equilibrio entre ambos.
